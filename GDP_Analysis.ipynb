```python
import os
```

```python
gsva=pd.DataFrame()
```

```python
path="C:\\Users\\DIPU\\Projects\\GDP Analysis\\Datasets\\GSVA\\"
```

```python
list_of_files=os.listdir(path=path)
```

```python
for filename in list_of_files:
    for statename in states:
        if statename in filename:
            filepath=path+filename
            data=pd.read_csv(filepath,encoding="ISO-8859-1")
            data=data[['Item','2014-15']]
            data=data.T
            data.columns=data.iloc[0]
            data=data[1:]
            data["state"]=statename
            gsva=pd.concat([data,gsva])
```

```python
gsva.head()
```

```python
gsva.set_index('state',inplace=True)
```

```python
gsva.head(3)
```

```python
gsva['Per Capita GSDP (Rs.)'].sort_values().plot.bar()
plt.show()
```

```python
gsva['Per Capita GSDP (Rs.)'].sort_values()['Goa']/gsva['Per Capita GSDP (Rs.)'].sort_values()['Bihar']
```

```python
gsva['Tertiary_gdp_percentage']=gsva['Tertiary']/gsva['Gross State Domestic Product']*100
gsva['Primary_gdp_percentage']=gsva['Primary']/gsva['Gross State Domestic Product']*100
gsva['Secondary_gdp_percentage']=gsva['Secondary']/gsva['Gross State Domestic Product']*100
```
